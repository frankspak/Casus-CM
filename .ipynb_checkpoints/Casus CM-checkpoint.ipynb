```python
# Find out what main genre is, then selects all movies with that main genre.
genres = dataset_1['genres'].values.tolist()
maingenres = []
for x in genres:
    if '|' in x:
        x = x.split('|')
        x = x[0]
    maingenres.append(x)


dataset_1['main_genre'] = maingenres
```

### Onderzoeksvraag 3 uitwerkingen.

```python
all_genres_duplicates = data_onderzoekvraag_3['main_genre'].values.tolist()
all_genres_once = []
for i in all_genres_duplicates: 
    if i not in all_genres_once: 
        all_genres_once.append(i) 
        
all_genres_once.sort()        
print(all_genres_once)
```

```python
all_years_duplicates = data_onderzoekvraag_3['title_year'].values.tolist()
all_years_once = []
for i in all_years_duplicates: 
    if i not in all_years_once:
        all_years_once.append(i) 
        
all_years_once = [round(x) for x in all_years_once]
all_years_once.sort()
print(all_years_once)
```

```python
uitslag = []
for i in range(len(all_genres_once)):
    genre = all_genres_once[i]
    for i in range(len(all_years_once)):
        year = all_years_once[i]
        for i in range(len(all_years_once)):
                    x = data_onderzoekvraag_3.loc[(data_onderzoekvraag_3['main_genre'] == genre) & (data_onderzoekvraag_3['title_year'] == year)]
                    avg = []
                    for i in range(len(x['imdb_score'])):
                        avg.append(i)  
                    if len(avg) > 0:
                        average = np.mean(x['imdb_score'])
                    else:
                        average = 0 # anders krijg je 'divide by zero' error
        #print('Het gemiddelde van het jaar {} voor het genre {} is: {}'.format(year, genre, average))
        out = (year, genre, average)
        uitslag.append(out)
        print(out)

print(uitslag)
```

```python
Uitslag_Dataset = pd.DataFrame(uitslag, columns =['Year', 'Genre', 'Average']) 
Uitslag_Dataset
```

```python
sns.scatterplot(x=Uitslag_Dataset["Average"], y=Uitslag_Dataset['Genre'], hue=Uitslag_Dataset['Year'])
```

# Casusopdracht: definitieve data science pipeline 

```python
# Obtain needed data from dataset
casus_opdracht = dataset_1.drop(['director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name', 'movie_facebook_likes', 'director_facebook_likes', 'cast_total_facebook_likes', 'num_user_for_reviews', 'genres', 'main_genre', 'title_year', 'content_rating', 'duration'], axis=1)
casus_opdracht.head()
```

```python
english_titles_score = []
non_english_titles_score = []

# Filter which movies are english or not and put them in their respective list
for index, row in casus_opdracht.iterrows():
    if row['language'] != "" and row['imdb_score'] != "": # make sure there is a variable
        if row['language'] == "English":
            english_titles_score.append(row['imdb_score'])
        else:
            non_english_titles_score.append(row['imdb_score'])

english_titles_score = random.sample(english_titles_score, 100)# take a random sample of 100

# Turn the scores into dataframes
datasheet_english = pd.DataFrame(english_titles_score, columns =['score']) 
datasheet_non_english = pd.DataFrame(non_english_titles_score, columns =['score']) 

display(datasheet_english, datasheet_non_english)
```

```python
#calculate average score of each list
average_english_score = sum(english_titles_score)/len(english_titles_score)
average_non_english_score = sum(non_english_titles_score)/len(non_english_titles_score)

# Combine the 2 scores in one dataframe
datasheet_scores = pd.DataFrame([(average_english_score, "English"), (average_non_english_score, "Non_English")], columns =['IMDB score', 'language']) 
datasheet_scores
```

```python
# plot the scores on a graph
sns.barplot(x='language', y='IMDB score', data=datasheet_scores)
```
